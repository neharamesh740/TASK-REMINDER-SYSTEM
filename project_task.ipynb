#Task reminder system

```python
import datetime

class Task:
    def __init__(self, name, deadline):
        self.name = name
        self.deadline = deadline

    def is_due_soon(self):
        today = datetime.date.today()
        days_left = (self.deadline - today).days
        return days_left <= 3 and days_left >= 0

def add_task(task_list, name, deadline_str):
    deadline = datetime.datetime.strptime(deadline_str, "%Y-%m-%d").date()
    task = Task(name, deadline)
    task_list.append(task)

def show_due_tasks(task_list):
    print("Tasks due within next 3 days:")
    for task in task_list:
        if task.is_due_soon():
            days_left = (task.deadline - datetime.date.today()).days
            print(f"- {task.name}, due in {days_left} day(s) on {task.deadline}")

# Example usage
tasks = []
add_task(tasks, "Submit Project Report", "2025-09-27")
add_task(tasks, "Client Meeting", "2025-09-29")
add_task(tasks, "Code Review", "2025-09-21")
add_task(tasks,"official meeting","2025-09-24")

show_due_tasks(tasks)
```
